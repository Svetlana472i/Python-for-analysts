## Task 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(10, 8))
corrMatrix = round(df.drop(columns=['id', 'date', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'yr_renovated']
).corr(method='spearman'), 2)

corrMatrix[abs(corrMatrix) < 0.3] = 0
sns.heatmap(corrMatrix, annot=True, linewidths=.5, cmap='RdYlGn')
plt.title("Матрица корреляции признаков")
plt.xticks(rotation=60);
```

Не  включаем в матрицу корреляции некоторые трудно формализуемые признаки 

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['condition'], df['price'])
plt.xlabel("Состояние")
plt.ylabel("Цена дома, млн")
plt.title("Зависимость цены дома от состояния");
```

Дома с лучшим состоянием продаются по более высокой цене, но данный показатель не является определяющим в формировании стоимости 

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg')
plt.xlabel("Жилая площадь, кв. м")
plt.ylabel("Цена дома, млн")
plt.title("Зависимость цены дома от жилой площади");
```

Наблбдается явно выраженная зависимость цены от величины жилой площади дома

```python
data = df.groupby('price')['bedrooms'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.scatter(data['bedrooms'], data['price'])
plt.xlabel("Количество спален")
plt.ylabel("Цена дома, млн")
plt.title("Зависимость цены дома от количества спален");
```

Прослеживается зависимость стоимости домов от количества спален

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['bathrooms'], y=df['price'], kind='reg')
plt.xlabel("Количество ванных")
plt.ylabel("Цена дома, млн")
plt.title("Зависимость цены дома от количества ванных");
```

Исключая крайние значения, прослеживается зависимость стоимости от количества ванных комнат

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['lat'], y=df['long'], kind='kde')
plt.xlabel("Широта")
plt.ylabel("Долгота")
plt.title("Зависимость количества предлагаемых домов от широты и долготы");
```

Есть предположение, что при широте от 47,5 до 47,8 находятся более густонаселенные районы, что может приводить к росту стоимости домов

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['lat'], y=df['price'], kind='reg')
plt.xlabel("Широта")
plt.ylabel("Стоимость")
plt.title("Зависимость стоимости дома от широты");
```

Подтверждается более высокая стоимость домов от 47,5 до 47,8 широты

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg')
plt.xlabel("Общая площадь дома")
plt.ylabel("Стоимость")
plt.title("Зависимость стоимости дома от общей площади дома");
```

Одно из самых сильных влияний на стоимость дома имеют его общая и жилая площади

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['sqft_lot'], y=df['price'], kind='reg')
plt.xlabel("Площадь земельного участка")
plt.ylabel("Стоимость")
plt.title("Зависимость стоимости дома от площади земельного участка");
```

Не подтверждается прямая зависимость стоимости дома от площади участка

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['waterfront'], y=df['price'], kind='reg')
plt.xlabel("Вид на море")
plt.ylabel("Стоимость")
plt.title("Зависимость стоимости дома от наличия вида на море");
```

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['view'], y=df['price'], kind='reg')
plt.xlabel("Вид из дома")
plt.ylabel("Стоимость")
plt.title("Зависимость стоимости дома от вида из дома");
```

Зависимость стоимости от вида на море и вида из дома не подтверждается, так как данный параметр необходимо исследовать для домов со схожими основными показателями
